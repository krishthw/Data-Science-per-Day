The table "messages data madeup.xlsx" contains send/receive message data for the application's users. Using Python, calculate what fraction of senders sent messages to at least 9 unique people on March 1, 2019. 

```python
import pandas as pd
data=pd.read_excel("messages data madeup.xlsx") # import excel file using pd.read_excel method
```

```python
march1data=data[data.date=='2019-03-01'] # extract information relevent to date 2019/03/01. 
#We also can use loc method >> march1data=data.loc[data['date']=='2019-03-01']
march1data
```

```python
march1data.dtypes 
```

```python
# group data by sender ID and Reciever ID. This will result a Groupby object
d=march1data.groupby(['sender ID','Receiver ID']) 
d
```

```python
# Let's check d
d.size()
```

If we groupby march1data by Sender ID and get the unique values of Reciever ID to a data frame, we can extract the information we want.

```python
# groupby march1data by Sender ID and get the unique values of Receiver ID to a data frame
Grouped=march1data.groupby('sender ID')['Receiver ID'].nunique().to_frame()
Grouped
```

```python
# Filter out senders who sent messages to more than 9 unique recievers 
filter1=Grouped.loc[Grouped['Receiver ID']>=9]
filter1
```
